# To prevent this exception handel should be done for any suspecious code







# There are 4 blocks :
### 1. try 
##### try for error
### 2. except
##### if error comes it will come to except block
### 3. else
##### if try block executed without any error then else will be excuted for residual code, else will not excecuted if try fails 
### 4. finally
##### it will be excuted at any situation 




```python
#sysntax

try:# try for error 
    1/0
except Exception as e: # if such error comes send it to except block
    print("This is Except Block::: ", e)
```

```python
try:  
    f=open('txt100.text', 'r')

except Exception as e:   #Exception is the supper class of all exeption 
    
    print("This is Except Block::: ", str(e).title())
```

```python
# else:

try:
    a=[]
    for i in range(0,10):
        if i<=0:
            a.append(0)  # we have not printed the list a here 
            
        else: 
            a.append(round(1/i,3)) 
                        
except Exception as e:
    print (str(e).capitalize())
else:
    print(a)
    print()
    print('Eception Handeled')
```

```python
# in this else will not be executed as try has not executed sucessfully
try:
    a=[]
    for i in range(0,10):
        
            a.append(1/i)
    
except Exception as e:
    print (f'There is a bug of : {str(e).title()}')
else:
    print(a)
    print()
    print('Eception Handeled')
```

```python
# finally 

try:
    a=[]
    for i in range(0,10):
        
            a.append(1/i)
    
except Exception as e:
    print (f'There is a bug of : {str(e).title()}')
else:
    print(a)
    print()
    print('Eception Handeled')
    
    
finally:
    print()
    print("FINALLY "*5) # it will be excecuted at any situation
```

# Custome Exception Handeling

##### Some time we need to raise custome msg for exception

### Example:


```python
# according to python the value for age is fine but not for you
age= int(input("ENTER YOUR AGE::::")) # enter age -111
print(age)



# to handel this kind of exception we need to create a sub class of Exception super class
```

```python
# creating a sub class of Exception class

class Validateage(Exception):
    def __init__(self, msg):  
        self.msg=msg
```

```python
# make and function to raise error

def validateage(age):
    if age<0:
        raise Validateage(f"Age can't be Negative {age} ")
    elif age>=200:
        raise Validateage(f"Age can't be too heigh  {age}")
    else:
        print(f"This is a Valid Age  {age}")
```

```python
# try without exception handelling method with age -111

age= int(input("ENTER YOUR AGE::::"))
validateage(age)
```

```python
# Custome Exception Handelling

try:
    age= int(input("ENTER YOUR AGE::::"))
    validateage(age)

except Validateage as v:
    print("There is a Bug as:::::", v)
else:
    print()
    age1=age*2
    print(f"Your age  will be double to {age1} in next {age} years")
```
